```python
# Metodo 1
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng
arr = np.arange(53)
N = len(arr)
M = 10
rest = N%M
arr[:-rest]
```

```python
k = int(N/M)
k
```

```python
def estimator(x):
    return np.mean(x**4)/(3*np.mean(x**2)**2)

arr_new = np.empty(N-rest)
rng = default_rng()
fake_estimator = np.empty(M)
xx = np.arange(M)
for j in range(M):
    for i in range(k):
        N_rnd = int(rng.uniform()*k)
        arr_new[i*M : (i+1)*M] = arr[ N_rnd*M : (N_rnd + 1) * M ]
        
    fake_estimator[j] = estimator(arr_new)
error = np.std(fake_estimator)
plt.scatter(xx,fake_estimator)
```

# Non linear beta array

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

beta_min = 0.36
beta_max = 0.50
N = 100
init_array = np.linspace(-1.2, 1.2, N)
plt.plot(init_array)
plt.title('Equispaziato')
plt.show()
trans = (np.tan(init_array))
trans = (trans + np.abs(np.min(trans)) )/(np.max(trans) - np.min(trans))
plt.plot(trans)
plt.title('Transformation')
plt.show()
beta_non_lin = trans * (beta_max - beta_min) + beta_min
# Non linear axis:
plt.scatter(beta_non_lin, 
            np.exp(-(beta_non_lin - 0.43)**2/(((beta_max - beta_min)/6)**2)), alpha = 0.4, s = 10)
plt.title('Non linear spacing for beta')
plt.show()
# Linear Axis
beta_lin = np.linspace(beta_min, beta_max, N)
plt.scatter(beta_lin, 
            np.exp(-(beta_lin - 0.43)**2/(((beta_max - beta_min)/6)**2)), alpha = 0.4, s = 10)
plt.title('Linear spacing for beta')
plt.show()
```
